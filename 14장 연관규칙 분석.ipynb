## 14장 연관규칙 분석

- 데이터 살펴보기

```python
#1. 모듈 및 함수 불러오기
import pandas as pd

#2. 데이터 불러오기
df = pd.read_csv('Retail_data.csv', engine='python', encoding='CP949')

#3. 상위 5개 행 불러오기
df.head()
```

- 데이터 전처리

```python
#1. 변수 선택
df = df.iloc[:,1:]

#2. 데이터 변환
df_dummy = pd.get_dummies(df)
df_dummy.head()
```

- Apriori 모형 기반 연관 규칙 생성

```python
#1. 모듈 및 함수 불러오기
from mlxtend.frequent_patterns import apriori, association_rules

#2. Apriori를 이용한 빈발항목세트 생성
frequent_items = apriori(df_dummy, min_support=0.1, use_colnames=True)
frequent_items.head()
```

```python
#1. 연관 규칙 생성
rules = association_rules(frequent_items, metric='confidence', min_threshold=0.6)

#2. 결과 출력
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].sort_values(by='lift', ascending=False).head(10)
```
